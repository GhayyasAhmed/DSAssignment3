# Question 1

## Add col "num_of_cast" in the dataset which contains the total number of cast in that particular movie or Tv-Show.

```python
# check if there is any null value in col "cast"
df["cast"].isnull().sum()
```

```python
# replace null value wit "None"

df["cast"] = df['cast'].replace(np.nan, "None")
```

```python
# total num of cast using this function

def get_no_of_cast(value):
    if value != "None":
        return len(value.split(", "))
    
    return 0
```

```python
# add col "num_of_cast" in the dataset

df["num_of_cast"] = df["cast"].apply(get_no_of_cast)
```
